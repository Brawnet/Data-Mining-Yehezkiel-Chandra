```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Mengimport Data menggunakan Pandas

```python
df = pd.read_csv('https://raw.githubusercontent.com/NathaliaMinoque/datasets/refs/heads/main/jabodetabek_house_price.csv')

df.head()
```

```python
# Calculate null values and percentage
null_counts = df.isnull().sum()
null_percentages = (null_counts / len(df)) * 100

# Create a DataFrame for columns with null values
null_df = pd.DataFrame({'Null Count': null_counts, 'Null Percentage': null_percentages})

# Filter for columns with null values
null_df = null_df[null_df['Null Count'] > 0]


# Format the percentage with two decimal places and % symbol
null_df['Null Percentage'] = null_df['Null Percentage'].map('{:.2f}%'.format)


null_df
```

UNTUK CLEANING DATA SAYA AKAN MENGGUNAKAN GLOBAL CONSTAN "UNKNOWN" KARENA DATA SEPERTI INI KALAU TIDAK DI CEK DI LAPANGAN DAN MELIHAT SERTIFIKAT AKAN SULIT UNTUK DI IDENTIFIKASI DENGAN PREDIKSI AI. TIDAK MENGGUNAKAN DELETING ROW KARENA DATA KOSONG DALAM JUMLAH YANG BANYAK

```python
# Global constants
MISSING_VALUE = "Unknown"
INVALID_VALUE = -1

df.fillna(MISSING_VALUE, inplace=True)
df = df.drop_duplicates()

df.isnull().sum()
```

MENCARI OUTLIER

```python
# Misalnya, kita akan memeriksa kolom 'price_in_rp' untuk mencari outlier
sns.boxplot(x=df['price_in_rp'])

# Menambahkan judul dan label
plt.title('Boxplot of Price in Rp')
plt.xlabel('Price in Rp')

# Menampilkan boxplot
plt.show()

# Menghitung nilai IQR (Interquartile Range) untuk mencari outlier
Q1 = df['price_in_rp'].quantile(0.25)
Q3 = df['price_in_rp'].quantile(0.75)
IQR = Q3 - Q1

# Menentukan batas bawah dan atas untuk outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Mencari outlier yang berada di luar rentang ini
outliers = df[(df['price_in_rp'] < lower_bound) | (df['price_in_rp'] > upper_bound)]
```

```python
df = df[df['price_in_rp'] <= 15000000000]  # Misalnya, kita tetapkan batas atas 20 miliar

# Menampilkan boxplot untuk harga yang sudah dibersihkan
sns.boxplot(x=df['price_in_rp'])
plt.title('Boxplot of Log-Transformed Price in Rp (Cleaned)')
plt.xlabel('Log Price in Rp')
plt.show()
```

UNTUK BINNING SAYA HANYA AKAN MENGGANTI LAND SIZE DAN BUILDING SIZE , LAND SIZE (< 50 mÂ², 50 - 100 mÂ², 100 - 200 mÂ², 200 - 500 mÂ², 500++ mÂ²), BUILDING SIZE (< 50 mÂ², 50 - 100 mÂ², 100 - 200 mÂ², 200 - 500 mÂ², 500++ mÂ²)

```python

# 1. Binning for Land Size
land_size_bins = [0, 50, 100, 200, 500, float('inf')]  # Define bins for land size
land_size_labels = ['< 50 mÂ²', '50 - 100 mÂ²', '100 - 200 mÂ²', '200 - 500 mÂ²', '500++ mÂ²']  # Labels for the bins

# Convert 'land_size_m2' to numeric, if there are any non-numeric values, they will be converted to NaN
df['land_size_m2'] = pd.to_numeric(df['land_size_m2'], errors='coerce')

# Apply binning for land size
df['land_size_bin'] = pd.cut(df['land_size_m2'], bins=land_size_bins, labels=land_size_labels, right=False)

# 2. Binning for Building Size
building_size_bins = [0, 50, 100, 200, 500, float('inf')]  # Define bins for building size
building_size_labels = ['< 50 mÂ²', '50 - 100 mÂ²', '100 - 200 mÂ²', '200 - 500 mÂ²', '500++ mÂ²']  # Labels for the bins

# Convert 'building_size_m2' to numeric, if there are any non-numeric values, they will be converted to NaN
df['building_size_m2'] = pd.to_numeric(df['building_size_m2'], errors='coerce')

# Apply binning for building size
df['building_size_bin'] = pd.cut(df['building_size_m2'], bins=building_size_bins, labels=building_size_labels, right=False)

# 3. Remove outliers for price (if necessary)
df_cleaned = df[df['price_in_rp'] < 2e11]  # Remove rows where price_in_rp is greater than 2e11

# 4. Display the cleaned DataFrame with selected columns
print(df_cleaned[['district', 'land_size_m2', 'land_size_bin', 'building_size_m2', 'building_size_bin', 'price_in_rp']].head())
```
